```python
from pyzillow.pyzillow import ZillowWrapper, GetDeepSearchResults
import pandas as pd
import csv
import pandas_profiling as pp
import matplotlib.pyplot as plt
import numpy as np
import requests
import itertools
from tqdm import tqdm

api_key = "X1-ZWz1fkoowi9erv_4dboj"
df = pd.read_csv('/Users/PatrickChen/Desktop/dsi-atx-1-projects-pchen2890/CAPSTONE/airbnb_data.csv',encoding="utf8")
```

```python
df.head()
```

```python
df.propertyType = df.propertyType.apply(lambda x: str(x))
df.house_num = df.house_num.apply(lambda x: str(x))
df.street_name = df.street_name.apply(lambda x: x.encode('utf8'))
```

```python
df["address"] = df.house_num+" "+df.street_name
```

```python
data = df[["propertyType","prices","zipcodes","address"]]
```

```python
data.address = data.address.str.replace(" ","+")
```

```python
data.head()
```

```python
home_type = []
tax_year = []
tax_value = []
year_built = []
property_size = []
home_size = []
bathrooms = []
bedrooms = []
zestimate_amount = []

for a,z in tqdm(itertools.izip_longest(data.address,data.zipcodes)):
    url = "http://www.zillow.com/webservice/GetDeepSearchResults.htm?zws-id="+api_key+"&address="+a+"&citystatezip="+str(z)
    response = requests.get(url)
    if "no exact match found" in str(response.text):
        home_type.append("none")
        tax_year.append("none")
        tax_value.append("none")
        year_built.append("none")
        property_size.append("none")
        home_size.append("none")
        bathrooms.append("none")
        bedrooms.append("none")
        zestimate_amount.append("none")
    else:
        zillow_data = ZillowWrapper(api_key)
        deep_search_response = zillow_data.get_deep_search_results(a, z)
        result = GetDeepSearchResults(deep_search_response)
        
        home_type.append(result.home_type)
        tax_year.append(result.tax_year)
        tax_value.append(result.tax_value)
        year_built.append(result.year_built)
        property_size.append(result.property_size)
        home_size.append(result.home_size)
        bathrooms.append(result.bathrooms)
        bedrooms.append(result.bedrooms)
        zestimate_amount.append(result.zestimate_amount)
```

```python
data["home_type"] = home_type
data["tax_year"] = tax_year
data["tax_value"] = tax_value
data["year_built"] = year_built
data["property_size"] = property_size
data["home_size"] = home_size
data["bathrooms"] = bathrooms
data["bedrooms"] = bedrooms
data["zestimate_amount"] = zestimate_amount
```

```python
data.head()
```

```python
data = data[data.zestimate_amount!="none"]
data = data.drop("address",axis=1)
```

```python
data.tax_year = data.tax_year.astype(float)
data.tax_value = data.tax_value.astype(float)
data.year_built = data.year_built.astype(float)
data.property_size = data.property_size.astype(float)
data.home_size = data.home_size.astype(float)
data.bathrooms = data.bathrooms.astype(float)
data.bedrooms = data.bedrooms.astype(float)
data.zestimate_amount = data.zestimate_amount.astype(float)
```
